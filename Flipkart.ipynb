```python
import requests as rq
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
url='https://www.flipkart.com/search?q=top%20xbox%20games&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
```

```python
req=rq.get(url)
```

```python
req
```

```python
soup=bs(req.text,'html.parser')
```

```python
output=pd.DataFrame(columns=['Product Name','Product For','Price'])
output
```

# Product name extract

```python



P1=soup.find_all('a',{'class':'wjcEIp'})
```

```python
p_name=[]
for text in P1:
    p_name.append(text.text)
    # print(text.text)
p_name
```

# extract the pricing

```python
P2=soup.find_all('div',{'class':"Nx9bqj"})
P2[1]
```

```python
p_price=[]
for i in P2:
    # if i==()
    p_price.append(i.text)
p_price=p_price[0:40]
```

```python
len(p_price)
# p_price
```

# extract discriptions

```python
P3=soup.find_all('div',{'class':"NqpwHC"})
```

```python
p_dis=[]
for j in P3:
    p_dis.append(j.text)
p_dis
```

# extract offers on product

```python
P4=soup.find_all('div',{'class':"UkUFwK"})
```

```python
p_offer=[]
for i in P4:
    p_offer.append(i.text)
```

```python
print(' offers',len(p_offer))
print(' product name',len(p_name)) 
print(' price',len(p_price))
print(' discription',len(p_dis))
```
